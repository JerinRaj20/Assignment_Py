```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
sub = pd.read_csv("sample_submission.csv")
```

```python
train_model = pd.read_csv("train_preprocessed.csv")
test_model = pd.read_csv("test_preprocessed.csv")
sample = pd.read_csv("sample_submission.csv")
```

```python
X = train_model.drop(columns = 'Response')
y = train_model['Response']
```

```python
train_X , test_X ,train_y , test_y = train_test_split(X,y,test_size =0.2, random_state =20)
```

```python
model = LinearRegression()
```

```python
model.fit(train_X, train_y)
```

```python
model.predict(test_X)
```

```python
model.coef_
model.intercept_
train_X_pred = model.predict(train_X)
test_X_pred = model.predict(test_X)
```

```python
train_X_RMSE = np.sqrt(mean_squared_error(train_y, train_X_pred))
```

```python
train_X_RMSE
```

```python
test_X_RMSE = np.sqrt(mean_squared_error(test_y, test_X_pred))
```

```python
test_X_RMSE
```

```python
train_X_R2 = r2_score(train_y, train_X_pred)
test_X_R2 = r2_score(test_y, test_X_pred)
```
